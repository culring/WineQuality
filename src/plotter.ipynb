```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_zlabel('z axis')
ax.scatter([1, 2], [1, 2], zs=[1, 2], label='label')
ax.set_xlim(10)
ax.set_ylim(10)
plt.legend()
```
